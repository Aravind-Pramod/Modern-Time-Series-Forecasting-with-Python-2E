```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

# Define the parameters
confidence_level = 0.90
alpha = 1 - confidence_level

# Calculate the critical values (quantiles)
lower_quantile = alpha / 2
upper_quantile = 1 - alpha / 2

# Generate a standard normal distribution
x = np.linspace(-3, 3, 1000)
y = stats.norm.pdf(x)

# Calculate the quantiles
lower_bound = stats.norm.ppf(lower_quantile)
upper_bound = stats.norm.ppf(upper_quantile)

# Plot the distribution
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution')
plt.fill_between(x, y, where=(x <= lower_bound), color='red', alpha=0.5, label='Lower Tail')
plt.fill_between(x, y, where=(x >= upper_bound), color='red', alpha=0.5, label='Upper Tail')
plt.fill_between(x, y, where=(x >= lower_bound) & (x <= upper_bound), color='green', alpha=0.5, label='90% Prediction Interval')

# Annotate the plot
plt.axvline(lower_bound, color='black', linestyle='--', label=f'Lower Quantile ({lower_quantile*100:.1f}th percentile)')
plt.axvline(upper_bound, color='black', linestyle='--', label=f'Upper Quantile ({upper_quantile*100:.1f}th percentile)')
plt.axvline(0, color='blue', linestyle='--', label='Mean (50th percentile)')

# Add labels and legend
# plt.title('Prediction Interval for a Standard Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend(loc='upper right')

plt.show()
```
